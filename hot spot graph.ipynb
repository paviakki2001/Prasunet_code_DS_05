```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Correctly read the CSV file (make sure the path is correct)
dataset = pd.read_csv(r"C:\Users\91936\Contacts\Downloads\archive (11)\NYC Accidents 2020.csv")
print(dataset.head())  # To display the first few rows of the dataset

# Setting a seed for reproducibility
np.random.seed(0)

# Generating random data
data = np.random.rand(10, 12)  # 10 rows and 12 columns

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title('Accident HOTSPOT')
plt.xlabel('accident ')
plt.ylabel('contribution')
plt.show()
```
